```python
def beautifulsoup_basic(url):
    return pd.read_html(url)[14][1:26].drop(['분석','Unnamed: 7'], axis=1)

def beautifulsoup_stock(url):
    return pd.read_html(url)[14][1:21][["기업명","공모가(원)","첫날종가(원)"]]
```

```python
def ipo_information(url):
    c_information = pd.read_html(url)
    ipo_name = c_information[16][1][0]
    market = c_information[16][1][1]
    category = c_information[16][1][2]
    sales = c_information[16][1][7]
    net_profit = c_information[16][1][8]
    capital = c_information[16][3][8]
    code = int(c_information[16][3][1])
    b_class = c_information[16][3][3]
    offical_price = c_information[17][1][3]
    bank = c_information[17][1][4]
    if c_information[18][0][0] == '인수회사':
        predict_demand = c_information[19][2][0]
        part_offering = c_information[19][2][1]
        ipo_date = c_information[19][2][5]
        c_r_o_i = c_information[19][1][8]
        lock_up = c_information[19][1][8]
    else:
        predict_demand = c_information[18][2][0]
        part_offering = c_information[18][2][1]
        ipo_date = c_information[18][2][5]
        c_r_o_i = c_information[18][1][8]
        lock_up = c_information[18][1][8]
    lst = [ipo_name,market,category,sales,net_profit,capital,code,b_class,offical_price,bank,predict_demand,part_offering,ipo_date,c_r_o_i,lock_up]
    return lst
```

```python
IPO_list_revise = []
```

```python
# 한국 IPO 정보 crawling
for loop in range(1,55,1):
    for value in range(1,2):
        # IPO 기업 page enter
        driver.find_element_by_xpath(f'/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[4]/tbody/tr[2]/td/table/tbody/tr[{value}]/td[1]/a/font').click()
        time.sleep(7)
#         try:
        # IPO 기업 이름
        IPO_name = driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[2]/tbody/tr[1]/td[2]').text.lstrip()
        # 종목코드
        IPO_code = int(driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[2]/tbody/tr[2]/td[4]').text.lstrip())
        # 주관사
        IPO_wnrkstk = driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[4]/tbody/tr[5]/td[2]').text.lstrip()
        # 수요예측
        tndydPcmr = driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[7]/tbody/tr[1]/td[3]').text.lstrip()
        # 공모청약
        rhdahcjddir = driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[6]/tbody/tr[2]/td[2]').text.lstrip()
        # 확정공모가
        rhdahrk = int(driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[4]/tbody/tr[4]/td[2]').text.lstrip()[:-1].rstrip().replace(',',''))
        # 공모주 상장일
        tkdwkddlf = driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[6]/tbody/tr[6]/td[2]').text.lstrip()
        # 의무보호확약
        ghkrdir = driver.find_element_by_xpath('/html/body/table[3]/tbody/tr/td/table[1]/tbody/tr/td[1]/table[6]/tbody/tr[9]/td[2]/table/tbody/tr/td[4]').text.lstrip().replace('%','')

        IPO_list_revise.append([IPO_name,IPO_code,IPO_wnrkstk,tndydPcmr,rhdahcjddir,rhdahrk,tkdwkddlf,ghkrdir])
        driver.back()
#         except:
#             driver.back()
#             continue
        time.sleep(7)
    URL = f'http://www.38.co.kr/html/fund/index.htm?o=k&page={loop+1}'
    driver.get(URL)
    
```
