```python
user_dict = {1:hansol_list, 2:dongjun_list, 3:hyunsoo_list, 4:kangwoo_list, 5:BingE_list}
```

```python
def insert_pr(user_dict):
    # get index number
    df = pd.DataFrame(index= image_list,columns = range(1,6)).fillna(0)
    for user, li_ in user_dict.items():
        for item in li_:
            try:
                ind = df.index.get_loc(item)
                df.iloc[ind,user-1] = 1
            except:
                continue
    return df
```

```python
sample = insert_pr(user_dict)
```

```python
sample.to_csv("prefer_data.csv", encoding='cp949')
```
